```python
import configparser
import postprocess
```

```python
sim_dir = "/scratch/tc427/lucy/simulation_SELEX/sim_results/sim_N20_long_hamming/"
```

```python
settings = configparser.ConfigParser()
settings.read(sim_dir+"settings.init")

aptamerSeq = settings.get('general', 'reference_aptamer')
roundNum = settings.getint('general', 'number_of_rounds')
outputFileNames = settings.get('general', 'experiment_name')
distanceMeasure = settings.get('selectionparams', 'distance')
```

```python
plt.style.use("seaborn-white")
fig, axes = plt.subplots(1, roundNum, figsize=(9, 30/roundNum), sharey=True)
postprocess.plot_histo_(roundNum, sim_dir+outputFileNames, aptamerSeq, distanceMeasure, axes)
fig.tight_layout()
```
